```python
import os

files = os.listdir("/Users/spring")
for f in files:
    print(f)
```

```python
def dumpdir(path):
    files = os.listdir(path)
    for f in files:
        fullpath = path + "/" + f
        if os.path.isdir(fullpath): #폴더이면 True
            print("[" + fullpath + "]")
            dumpdir(fullpath) #재귀함수 (for문 반복)
        else:
            print("/t" + fullpath) #file이면 False
```

```python
dumpdir("/Users/spring")
```

```python
path = "/Users/spring/test"
files = os.listdir(path)
for f in files:
    if (f.find("-") and f.endswith(".md")):
        name = f[0:-3] #확장자 빼고 출력 (슬라이싱 실수 줄이기)
        print(name)
        ext = f[-3:] #확장자 출력
        print(ext)
        part = name.split("-") # '-' 기준으로 분리 
        newname = part[1].strip() + "-" + part[0].strip() + ext #앞,뒤 내용 바꾸기 
        #strip() : 공백제거
        print(newname)
        os.rename(path + "/" + f, path + "/" + newname)
        
```
